```python
import os
import json
import matplotlib.pyplot as plt

def analizza_json(directory_path):
    table_counts = []

    # Itera su tutti i file JSON nella cartella
    for file_name in sorted(os.listdir(directory_path)):
        if file_name.endswith('.json'):
            with open(os.path.join(directory_path, file_name), 'r', encoding='utf-8') as f:
                data = json.load(f)
                
                # Conta le tabelle nel file
                num_tables = len(data)
                table_counts.append(num_tables)

    # Grafico con il numero di tabelle per JSON
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, len(table_counts) + 1), table_counts, marker='o', linestyle='-', color='red', label='Numero di tabelle')
    plt.title('Numero di tabelle per file JSON')
    plt.xlabel('Progressivo JSON')
    plt.ylabel('Numero di tabelle')
    plt.grid(True)
    plt.xticks(range(1, len(table_counts) + 1))  # Mostra solo numeri interi sull'asse X
    plt.tight_layout()
    plt.legend()
    plt.show()

    # Ritorna i dati per ulteriori analisi se necessario
    return table_counts

# Esempio di utilizzo
directory_path = 'extraction_RAG'
analizza_json(directory_path)
```
